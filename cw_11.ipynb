Exercise 11: Creating Line Grapgh Using Different Libraries
Let's compare the creation process between Matplotlib, Pandas, and Seaborn. We will create a
Pandas DataFrame with random values and plot it using various methods.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
X = np.arange(0,100)
Y = np.random.randint(0,200, size=X.shape[0])
```

```python
plt.plot(X, Y)
```

```python
df = pd.DataFrame({'x':X, 'y_col':Y})
```

```python
df.plot('x', 'y_col')
```

```python
sns.lineplot(data=df, x=X, y=Y)
```
