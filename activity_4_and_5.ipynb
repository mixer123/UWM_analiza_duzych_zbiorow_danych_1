```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'cars.data'
df = pd.read_csv(url, delim_whitespace=True)

column_names = ['mpg', 'cylinders', 'displacement [cu]','horsepower','weight', 'acceleration', 'year', 'origin', 'name']
column_names
len(df.columns)
```

```python
df
```

```python
df.columns = ['mpg', 'cylinders', 'displacement [cu]','horsepower','weight', 'acceleration', 'year', 'origin', 'name']
```

```python
df['horsepower']
```

```python
df = df.replace('?',0)
```

```python
df['horsepower']=pd.to_numeric(df['horsepower'])
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=df['horsepower'], y=df['weight'])
plt.show()
```

```python
df.horsepower.plot(kind='hist')
```

```python
sns.distplot(df['horsepower'], bins=10)
```

```python
df.boxplot(column='horsepower', by='cylinders')
```

```python
fig, ax = plt.subplots()
df.groupby('year')['horsepower'].mean().plot(ax=ax,
label='horsepower')
ax.legend()
```

```python
fig, ax = plt.subplots()
df.groupby('year')['horsepower','acceleration','mpg'].mean().plot(ax=ax,
label='horsepower')
plt.legend([df.horsepower, df.mpg])
```
