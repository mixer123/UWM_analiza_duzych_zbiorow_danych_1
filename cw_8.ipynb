```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X =[]
Y =[]
Y1 = []

for i in range(-4,5):
    X.append(i + 0.02)
    Y.append(4 * i ** 2)
    Y1.append( 4 * i**4 -12 * i**2 +20)

fig, ax = plt.subplots()
ax.plot(X, Y)
ax.plot(X,Y1)

plt.show()
```

```python
X = np.linspace(-4,4,100)
Y = np.power(X,2)
Y1 = np.power(X,4)  -12 * np.power(X,2) +20
fig, ax = plt.subplots()
ax.plot(X, Y)
ax.plot(X,Y1)

plt.show()
```
